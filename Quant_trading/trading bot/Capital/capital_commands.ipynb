```python
def account_balance(demokeys, acount_name):
    # Establish a connection to the demo API server
    conn = http.client.HTTPSConnection("demo-api-capital.backend-capital.com")
    # Define an empty payload for the HTTP GET request
    payload = ''
    # Set the headers with X-SECURITY-TOKEN and CST values
    headers = {
        'X-SECURITY-TOKEN': demokeys[1],
        'CST': demokeys[0]
    }
    # Send an HTTP GET request to retrieve account information
    conn.request("GET", "/api/v1/accounts", payload, headers)
    # Get the response from the server
    res = conn.getresponse()
    # Read the data from the response
    data = res.read()
    # Parse the JSON data received from the server
    json_object = json.loads(data)
    # Iterate through the accounts in the JSON response
    for i in range(len(json_object['accounts'])):
        # Check if the accountName matches the specified acount_name
        if json_object['accounts'][i]['accountName'] == acount_name:
            # Return the balance of the matched account
            return json_object["accounts"][i]['balance']['balance']
```

```python
def acount_balance(demokeys,acount_name):
    conn = http.client.HTTPSConnection("demo-api-capital.backend-capital.com")
    payload = ''
    headers = {
      'X-SECURITY-TOKEN': demokeys[1],
      'CST': demokeys[0]
    }
    conn.request("GET", "/api/v1/accounts", payload, headers)
    res = conn.getresponse()
    data = res.read()
    json_object = json.loads(data)
    for i in range(len(json_object['accounts'])):
        if(json_object['accounts'][i]['accountName'] == acount_name):
            return json_object["accounts"][i]['balance']['balance']
```
